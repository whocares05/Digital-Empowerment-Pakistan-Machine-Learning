# Important Function
# OpenCV (`cv2`) Functions Overview


## 1. `cv2.imread()`
**Purpose**: Load an image from a file.

**Parameters**:
- `filename`: Path to the image file.
- `flags`: Image loading flags (e.g., `cv2.IMREAD_GRAYSCALE`, `cv2.IMREAD_COLOR`).

**Example**:
```python
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
```
##  2. `cv2.resize()`

**Purpose:** Resize an image.

Parameters:

- **src:** Input image.
- **dsize:** Output size as (width, height).
- **fx:** Scale factor along the horizontal axis (optional).
- **fy:** Scale factor along the vertical axis (optional).
- **interpolation:** Interpolation method (e.g., cv2.INTER_LINEAR).

```python
resized_image = cv2.resize('image.jpg', (width, height), interpolation=cv2.INTER_LINEAR)
```

## 3. `cv2.GaussianBlur()`

**Purpose**: Apply Gaussian blur to an image to reduce noise and detail. This is often used for smoothing images.

**Description**: The `cv2.GaussianBlur()` function applies a Gaussian filter to the image, which blurs the image and reduces its details. This can be useful for noise reduction or smoothing before applying other image processing techniques.

**Parameters**:
- **`src`**: Input image.
  - **Type**: `numpy.ndarray`
- **`ksize`**: Kernel size as `(width, height)`. Both dimensions must be odd numbers. The size of the kernel affects the extent of the blur.
  - **Type**: `tuple`
  - **Example**: `(5, 5)`
- **`sigmaX`**: Standard deviation in the X direction. A larger value results in a more pronounced blur.
  - **Type**: `float`
  - **Example**: `0`

**Example**:
```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```
## 4.  `cv2.filter2D()`

**Purpose**: Apply a custom convolutional filter to an image. This allows for advanced filtering operations like sharpening or edge detection.

**Description**: The `cv2.filter2D()` function convolves an image with a kernel. This kernel can be used to perform various image processing tasks such as sharpening, blurring, or edge detection based on the kernel values.

**Parameters**:
- **`src`**: Input image.
  - **Type**: `numpy.ndarray`
- **`ddepth`**: Depth of the output image. Typically set to `-1` to use the same depth as the source image.
  - **Type**: `int`
  - **Example**: `-1`
- **`kernel`**: Convolution kernel. A matrix that defines the filter.
  - **Type**: `numpy.ndarray`

**Example**:
```python
filtered_image = cv2.filter2D(image, -1, kernel)
```
## 5.  `cv2.threshold()`

**Purpose**: Apply a fixed-level threshold to an image, which is useful for image segmentation and binarization.

**Description**: The `cv2.threshold()` function converts an image to a binary image by applying a threshold. Pixels with values above the threshold are set to a maximum value, while others are set to 0 or another specified value. This is commonly used for image segmentation tasks.

**Parameters**:
- **`src`**: Input image.
  - **Type**: `numpy.ndarray`
- **`thresh`**: Threshold value. Pixels with a value greater than this will be set to `maxval`.
  - **Type**: `float`
- **`maxval`**: Maximum value to use with the `THRESH_BINARY` and `THRESH_BINARY_INV` thresholding types.
  - **Type**: `float`
- **`type`**: Thresholding type. Examples include `cv2.THRESH_BINARY` for binary thresholding or `cv2.THRESH_BINARY_INV` for inverted binary thresholding.
  - **Type**: `int`
  - **Example**: `cv2.THRESH_BINARY`

**Example**:
```python
thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```
## 6.  `cv2.Canny()`

**Purpose**: Apply the Canny edge detection algorithm to identify edges in an image. This is useful for detecting boundaries and features.

**Description**: The `cv2.Canny()` function performs edge detection using the Canny algorithm. This method identifies edges by detecting areas of the image with a high gradient, which correspond to boundaries between different regions.

**Parameters**:
- **`image`**: Input image.
  - **Type**: `numpy.ndarray`
- **`threshold1`**: Lower threshold for the hysteresis procedure. Edge pixels with gradient values higher than `threshold2` are considered strong edges.
  - **Type**: `float`
- **`threshold2`**: Upper threshold for the hysteresis procedure. Pixels with gradient values between `threshold1` and `threshold2` are considered weak edges.
  - **Type**: `float`

**Example**:
```python
edges = cv2.Canny(image, 100, 200)
```

## 7.   `cv2.HoughCircles()`

**Purpose**: Detect circles in an image using the Hough Circle Transform. This is useful for finding circular shapes in images.

**Description**: The `cv2.HoughCircles()` function detects circles in an image by using the Hough Circle Transform. It identifies circles based on the parameters provided, such as the minimum distance between circles and the accumulator thresholds.

**Parameters**:
- **`image`**: Input image.
  - **Type**: `numpy.ndarray`
- **`method`**: Detection method. Typically `cv2.HOUGH_GRADIENT`.
  - **Type**: `int`
- **`dp`**: Inverse ratio of the accumulator resolution to the image resolution. A value of 1 means the accumulator has the same resolution as the image.
  - **Type**: `float`
- **`minDist`**: Minimum distance between the centers of the detected circles.
  - **Type**: `float`
- **`param1`**: First method-specific parameter, such as the higher threshold for the Canny edge detector.
  - **Type**: `float`
- **`param2`**: Second method-specific parameter, such as the accumulator threshold for the circle centers.
  - **Type**: `float`

**Example**:
```python
circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=30)
```

# Convolution
**Convolution** is a mathematical operation used to blend two functions. In the context of image processing, convolution is applied to an image with a filter (or kernel) to extract features such as edges, blur, or sharpen the image.

The process involves moving the kernel over the image, pixel by pixel, and at each position, performing an element-wise multiplication between the kernel values and the corresponding image pixel values. The sum of these multiplications gives the output pixel value for that position.

## Example in Image Processing:
For example, a **3x3** kernel can be used to detect edges in an image. The kernel slides over the image, multiplying its values with the corresponding pixel values and then summing the result to produce a single pixel in the output image.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/sample_data/parrot.jpg', 0)

# Define an edge detection kernel (Sobel operator)
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# Apply the convolution using OpenCV's filter2D function
convoluted_image = cv2.filter2D(image, -1, kernel)

# Use matplotlib to display the original and convoluted images
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image,)
plt.title('Original Image')
plt.axis('off')

# Display edge-detected image (convoluted image)
plt.subplot(1, 2, 2)
plt.imshow(convoluted_image)
plt.title('Edge Detected Image')
plt.axis('off')

# Show both images
plt.tight_layout()
plt.show()

````

# Gray Scalling
Gray-scaling is a technique in image processing where a color image is converted into a grayscale (or black-and-white) image. This process reduces the complexity of an image by removing color information, leaving only the intensity of light. Each pixel in a grayscale image represents a shade of gray, ranging from 0 (black) to 255 (white).

Why Gray-scaling?
Simplification: Gray-scaling reduces the complexity of the image, making it easier to process for tasks like edge detection or object recognition.
Efficiency: Processing grayscale images is faster as they require less computational resources compared to colored images.
Preprocessing Step: Many machine learning models and computer vision algorithms work better with grayscale images.

````python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1, 2, figsize=(16, 8))
fig.tight_layout()

ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original")

ax[1].imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
ax[1].set_title("Grayscale")
plt.show()
````

### **Image Translation refers to shifting an image along the x and y axes. This operation can be useful for tasks like object detection, data augmentation, or repositioning image elements.**

In OpenCV, you can translate an image using the `cv2.warpAffine()` function and a translation matrix.
Implement image translation using OpenCV with the following steps:
- **Define the Translation Matrix:** Create a translation matrix to shift the image by `100` pixels along the `x-axis` and `50` pixels along the `y-axis`.

- **Get Image Dimensions:** Extract the number of rows and columns from the image.
Apply the Translation: Use` cv2.warpAffine` to apply the translation matrix to the image.
Convert the Image for Display: Convert the translated image from BGR to RGB format.

#Task 3: Image Rotation using OpenCV

**Image Rotation** refers to rotating an image by a certain angle around a specific center point, often the image's center. OpenCV provides the `cv2.warpAffine()` function combined with a rotation matrix to achieve this.

---

### **Steps to Implement Image Rotation:**

- **Obtain the Image Dimensions**:
  - Extract the **height** and **width** of the image.

- **Define the Rotation Center**:
  - Set the center of the image around which the rotation will occur (often the center).

- **Copy the Image Four Times**:
  - Create four copies of the original image.
  - Rotate each image by **90**, **180**, **270**, and **360 degrees**.

- **Specify the Rotation Angle**:
  - Rotate the image counter-clockwise by the required degrees.
  
- **Get the Rotation Matrix**:
  - Use `cv2.getRotationMatrix2D()` to compute the matrix needed for the rotation.

- **Perform the Rotation**:
  - Apply the rotation matrix to the image using `cv2.warpAffine()`.

- **Convert the Image for Display**:
  - Convert the rotated image from **BGR to RGB** format.

- **Display the Rotated Image**:
  - Use **matplotlib** to display the rotated images without axis labels.

---





````python
# Check if the image is loaded correctly
if image is None:
    raise ValueError("Image could not be loaded. Please check the file path.")

# Get the dimensions of the image
height, width = image.shape[:2]

# Define the center of rotation (center of the image)
center = (width // 2, height // 2)

# Function to rotate the image by a specific angle
def rotate_image(image, angle, center):
    # Get the rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    # Perform the rotation
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    # Convert to RGB format for Matplotlib
    rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)
    return rotated_image_rgb

# Rotate the image by 90, 180, 270, and 360 degrees
angles = [90, 180, 270, 360]
rotated_images = [rotate_image(image, angle, center) for angle in angles]

# Create a figure to display the rotated images
fig, ax = plt.subplots(1, 4, figsize=(20, 5))

# Display each rotated image
for i, rotated_image in enumerate(rotated_images):
    ax[i].imshow(rotated_image)
    ax[i].set_title(f'Rotation {angles[i]}')
    ax[i].axis('off')  # Hide the axis labels

# Show the plot
plt.tight_layout()
plt.show()

print("Task 3")
````

### **Lab Instructions:**

Your task is to perform image scaling and resizing using **OpenCV** and **Matplotlib**. Follow the steps below to demonstrate different scaling techniques.

---

### **Steps:**

1. **Load an Image**:
   - Use **OpenCV** to load an image.
   - Ensure the image is loaded correctly before proceeding with resizing operations.

2. **Scale the Image**:
   - Scale the image to **15%** of its original size using **linear interpolation**.

3. **Enlarge the Image**:
   - Enlarge the image to **twice its original size** using **cubic interpolation**.

4. **Resize to Specific Dimensions**:
   - Resize the image to specific dimensions of **200x400 pixels** using **area interpolation**.

5. **Visualize the Results**:

````python


def enhance_brightness_contrast(image, alpha=1.5, beta=50):
    """Enhance the brightness and contrast of an image."""
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

def apply_gaussian_blur(image, kernel_size=(15, 15)):
    """Apply Gaussian blur to reduce noise."""
    return cv2.GaussianBlur(image, kernel_size, 0)

def detect_edges(image):
    """Detect edges using Canny edge detection."""
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray_image, 100, 200)
    return edges

def main():

    # Check if the image is loaded correctly
    if image is None:
        raise ValueError("Image could not be loaded. Please check the file path.")

    # Step 1: Enhance brightness and contrast
    enhanced_image = enhance_brightness_contrast(image)

    # Step 2: Apply Gaussian blur
    blurred_image = apply_gaussian_blur(enhanced_image)

    # Step 3: Detect edges
    edges_image = detect_edges(blurred_image)

    # Display the processed images using Matplotlib
    plt.figure(figsize=(15, 5))

    # Original Image
    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Enhanced Image
    plt.subplot(1, 4, 2)
    plt.imshow(cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB))
    plt.title('Enhanced Brightness & Contrast')
    plt.axis('off')

    # Blurred Image
    plt.subplot(1, 4, 3)
    plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
    plt.title('Gaussian Blurred Image')
    plt.axis('off')

    # Edges Image
    plt.subplot(1, 4, 4)
    plt.imshow(edges_image, cmap='gray')
    plt.title('Detected Edges (Canny)')
    plt.axis('off')

    # Show the plot
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()

print("Task 10")
````

# Task 12 Linear Spatial Filtering

**Description:**  
Linear spatial filtering is a technique used in image processing to modify an image based on the linear combination of pixel values in its neighborhood. This process involves using a filter (or kernel) that slides over the image to perform operations such as smoothing, sharpening, and edge detection.

---

### **Key Concepts:**

1. **Filter (Kernel):**  
   - A small matrix (e.g., 3x3, 5x5) used to process the image. Each element in the kernel represents a weight that modifies the corresponding pixel in the image.

2. **Convolution:**  
   - The core operation in linear spatial filtering. It involves overlaying the filter on the image, multiplying the filter values with the corresponding pixel values, summing the results, and replacing the central pixel with this sum.

3. **Filtering Operations:**  
   - **Smoothing (Blurring):**  
     Reduces noise and detail by averaging pixel values, which results in a softening effect. Common kernels include the average filter and Gaussian filter.
   - **Sharpening:**  
     Enhances edges and fine details by emphasizing differences between pixel values. Common kernels include the Laplacian filter and the unsharp mask.
   - **Edge Detection:**  
     Highlights the boundaries of objects within an image. Filters like the Sobel and Prewitt operators are commonly used for this purpose.

---

````python

def apply_filter(image, kernel):
    """Apply the given filter (kernel) to the image using convolution."""
    return cv2.filter2D(image, -1, kernel)

def main():
    # Load the image
    image_path = 'path_to_your_image.png'  # Replace with your image path
    image = cv2.imread(image_path)

    # Check if the image is loaded correctly
    if image is None:
        raise ValueError("Image could not be loaded. Please check the file path.")

    # Define kernels for filtering operations

    # Smoothing (Blurring) Kernel (Average Filter)
    avg_kernel = np.ones((5, 5), np.float32) / 25  # 5x5 average kernel

    # Sharpening Kernel (Unsharp Mask)
    sharp_kernel = np.array([[0, -1, 0],
                             [-1, 5, -1],
                             [0, -1, 0]])  # 3x3 sharpening kernel

    # Edge Detection Kernel (Sobel Operator)
    sobel_kernel_x = np.array([[-1, 0, 1],
                                [-2, 0, 2],
                                [-1, 0, 1]])  # Sobel X kernel
    sobel_kernel_y = np.array([[1, 2, 1],
                                [0, 0, 0],
                                [-1, -2, -1]])  # Sobel Y kernel

    # Apply the filters
    blurred_image = apply_filter(image, avg_kernel)
    sharpened_image = apply_filter(image, sharp_kernel)
    edges_x = apply_filter(image, sobel_kernel_x)
    edges_y = apply_filter(image, sobel_kernel_y)

    # Combine Sobel X and Y to get the magnitude of edges
    edges = cv2.magnitude(edges_x, edges_y)

    # Display the results
    plt.figure(figsize=(15, 10))

    # Original Image
    plt.subplot(2, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Blurred Image
    plt.subplot(2, 3, 2)
    plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
    plt.title('Blurred Image (Average Filter)')
    plt.axis('off')

    # Sharpened Image
    plt.subplot(2, 3, 3)
    plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
    plt.title('Sharpened Image (Unsharp Mask)')
    plt.axis('off')

    # Edges X
    plt.subplot(2, 3, 4)
    plt.imshow(edges_x, cmap='gray')
    plt.title('Edges (Sobel X)')
    plt.axis('off')

    # Edges Y
    plt.subplot(2, 3, 5)
    plt.imshow(edges_y, cmap='gray')
    plt.title('Edges (Sobel Y)')
    plt.axis('off')

    # Combined Edges
    plt.subplot(2, 3, 6)
    plt.imshow(edges, cmap='gray')
    plt.title('Combined Edges (Magnitude)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()

print("Task 12")
````
