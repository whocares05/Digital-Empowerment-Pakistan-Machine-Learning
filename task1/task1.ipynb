```python
df = pd.read_csv('zameen-updated.csv')
import pandas as pd
from datetime import datetime
df['date_added'] = pd.to_datetime(df['date_added'], format='%m-%d-%Y', errors='coerce')

reference_date = datetime(2024, 1, 1)


def calculate_month_difference(start_date, end_date):
    if pd.isnull(start_date):
        return None
    return (end_date.year - start_date.year) * 12 + end_date.month - start_date.month
    df['age'] = df['date_added'].apply(lambda x: calculate_month_difference(x, reference_date))

df.head()
```

```python
numerical=df.describe()
numerical.columns
numerical=df[numerical.columns]
numerical.median()
```

```python
df.describe()
```

```python
df.drop(columns=['property_id', 'location_id', 'page_url','location','latitude','longitude','date_added','agency','agent'], inplace=True)
df.head()
```

```python
def convert_area(value):
    number, unit = value.split()
    number = float(number.replace(',', ''))
    if unit == 'Marla':
        return number * 1
    elif unit == 'Kanal':
        return number * 20
    else:
        return None


df['area'] = df['area'].apply(convert_area)
df.head()
```

```python
df.drop(columns=['Area Type','Area Size','Area Category'], inplace=True)
df.head()
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns


df_encoded = pd.get_dummies(df, columns=categorical_columns)
df_encoded.columns
```

```python
df_encoded.drop(columns=['property_type_Farm House','city_Faisalabad','province_name_Islamabad Capital','purpose_For Rent'], inplace=True)
```

```python
df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = df_encoded.drop(columns=['price'])
target = df_encoded['price']
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_columns = features.select_dtypes(include=['int', 'float']).columns
features[numerical_columns] = scaler.fit_transform(features[numerical_columns])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
